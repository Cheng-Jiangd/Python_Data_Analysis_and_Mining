```python
import pandas as pd
import numpy as np
```

```python
bc_data = pd.read_csv('bc_data.csv', header=0)
bc_data.head()
```

```python
bc_data.shape
```

```python
print(bc_data.columns)
```

```python
bc_data.describe()
```

```python
bc_data.info()
```

```python
data = bc_data.drop(['id'], axis=1)

print(data.head(2))
```

```python
X_data = data.drop(['diagnosis'], axis=1)
X_data.head(2)
```

```python
from sklearn.preprocessing import StandardScaler
means_norm = StandardScaler()
means_norm.fit(X_data)
X_data = means_norm.transform(X_data)
X_data[2]
```

```python
y_data = np.ravel(data[['diagnosis']])
y_data[0:6]
```

```python
from sklearn.model_selection import train_test_split
X_trainSet, X_testSet, y_trainSet, y_testSet = train_test_split(X_data,
                                                                y_data,
                                                                test_size=0.25,
                                                                random_state=1)
```

```python
print(X_trainSet.shape)
```

```python
print(X_testSet.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
myModel = KNeighborsClassifier(algorithm='kd_tree')
```

```python
help(myModel)
```

```python
myModel.fit(X_trainSet, y_trainSet)
```

```python
myModel.get_params()
```

```python
y_predSet = myModel.predict(X_testSet)
```

```python
y_predSet
```

```python
y_testSet
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
print(accuracy_score(y_testSet, y_predSet))
```

```python
precision = precision_score(y_testSet, y_predSet, pos_label='M')
recall = recall_score(y_testSet, y_predSet, pos_label='M')
precision, recall
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
y_pred_proba = myModel.predict_proba(X_testSet)[:, 1]

fpr, tpr, thresholds = roc_curve(y_testSet, y_pred_proba, pos_label='M')
roc_auc = auc(fpr, tpr)

# ç»˜åˆ¶ROCæ›²çº¿
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
NumOfNeighbors = range(1, 23)
KNNs = [KNeighborsClassifier(n_neighbors=k) for k in NumOfNeighbors]
range(len(KNNs))
```

```python
scores = [KNNs[i].fit(X_trainSet, y_trainSet).score(X_testSet, y_testSet) for i in range(0, 22)]
scores
```

```python
import matplotlib.pyplot as plt
plt.plot(NumOfNeighbors, scores)
plt.xlabel('Number of Neighbors')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.xticks(NumOfNeighbors)
plt.show()
```
