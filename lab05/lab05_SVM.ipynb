```python
# 导入必要的库
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# 加载数据
bc_data = pd.read_csv('bc_data.csv', header=0)
data = bc_data.drop(['id'], axis=1)
X_data = data.drop(['diagnosis'], axis=1)
y_data = np.ravel(data[['diagnosis']])
```

```python
# 数据标准化
means_norm = StandardScaler()
means_norm.fit(X_data)
X_data = means_norm.transform(X_data)
```

```python
# 划分数据集
X_trainSet, X_testSet, y_trainSet, y_testSet = train_test_split(X_data, y_data, test_size=0.25, random_state=1)
```

```python
# # 创建SVM模型 with default regularization
# svm_model = SVC(kernel='linear')
```

```python
# # SVM with polynomial kernel
# svm_model = SVC(kernel='poly', C=0.2)

# # SVM with RBF kernel
# svm_model = SVC(kernel='rbf', C=0.2)

# SVM with sigmoid kernel
svm_model = SVC(kernel='sigmoid', C=0.2)
```

```python
# 训练模型
svm_model.fit(X_trainSet, y_trainSet)
```

```python
# 预测测试集
y_pred_svm = svm_model.predict(X_testSet)
```

```python
# 计算准确率
accuracy_svm = accuracy_score(y_testSet, y_pred_svm)
print(accuracy_svm)
```
