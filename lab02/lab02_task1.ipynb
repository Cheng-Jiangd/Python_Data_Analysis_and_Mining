- 填充缺失值

```python
rd = pd.read_excel("./height_weight_none.xlsx")
rd
```

```python
rd = rd.replace(to_replace='nul', value=np.nan)
rd
```

```python
a = rd.fillna(100)
a
```

```python
b = rd.fillna(method='ffill', axis=0)
b
```

```python
c = rd.fillna(method='bfill', axis=1)
c
```

```python
d = rd.fillna(rd.mean())
d
```
